```python
import easyocr
import cv2
import os
import numpy as np
```

```python
# Load your trained YOLOv8 model
yolo_model_path = '/content/drive/MyDrive/LPR_YOLO_Folder/runs/detect/train/weights/best.pt'
yolo_model = YOLO(yolo_model_path)

# Initialize the EasyOCR reader (English assumed, add other languages if needed)
reader = easyocr.Reader(['en'])
```

```python
#  Process Images
def process_image(image_path, output_dir):
    # Create output directory if it doesn't exist
    os.makedirs(output_dir, exist_ok=True)

    # Read the input image
    image = cv2.imread(image_path)

    # Detect license plates using YOLOv8
    results = yolo_model(image)
    detections = results[0].boxes.data.cpu().numpy()

    # Iterate through detected bounding boxes
    for i, box in enumerate(detections):
        x1, y1, x2, y2, conf, cls = box.astype(int)

        # Crop the license plate region
        cropped_plate = image[y1:y2, x1:x2]

        # Save the cropped plate image
        cropped_plate_path = os.path.join(output_dir, f'cropped_plate_{i}.jpg')
        cv2.imwrite(cropped_plate_path, cropped_plate)

        # Recognize text from the cropped plate using EasyOCR
        text_results = reader.readtext(cropped_plate)
        for (bbox, text, prob) in text_results:
            print(f"Detected Text: {text} (Confidence: {prob:.2f})")

            # Annotate the image with bounding box and text
            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(image, text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Save the annotated image
    annotated_image_path = os.path.join(output_dir, 'annotated_image.jpg')
    cv2.imwrite(annotated_image_path, image)

    print(f"Processed image saved at: {annotated_image_path}")

    # Path to input image
image_path = '/content/sample_image.jpg'

# Directory to save output results
output_dir = '/content/output_images'

# Process the image
process_image(image_path, output_dir)
```

```python
# video process

def process_video(video_path, output_video_path):
    # Open the video file
    cap = cv2.VideoCapture(video_path)

    # Get video properties
    width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps = cap.get(cv2.CAP_PROP_FPS)

    # Define video writer
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

    # Process each frame
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Detect license plates in the frame
        results = yolo_model(frame)
        detections = results[0].boxes.data.cpu().numpy()

        for box in detections:
            x1, y1, x2, y2, conf, cls = box.astype(int)

            # Crop the license plate region
            cropped_plate = frame[y1:y2, x1:x2]

            # Recognize text from the cropped plate using EasyOCR
            text_results = reader.readtext(cropped_plate)
            for (bbox, text, prob) in text_results:
                print(f"Frame Text Detected: {text} (Confidence: {prob:.2f})")

                # Annotate the frame with bounding box and text
                cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                cv2.putText(frame, text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

        # Write the annotated frame to the output video
        out.write(frame)

    # Release resources
    cap.release()
    out.release()

    print(f"Processed video saved at: {output_video_path}")

# Path to input video
video_path = '/content/drive/MyDrive/dataset/Object_detec_task/Vehicles Driving Through Flooding A93 Road Perthshire Scotland.mp4'

# Path to save output video
output_video_path = '/content/output_video.mp4'

# Process the video
process_video(video_path, output_video_path)

```
